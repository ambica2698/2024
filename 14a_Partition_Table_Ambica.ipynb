#### Vertical and Horizontal partitioning  
**Vertical partitioning**: some columns are moved to new tables. Each table contains the same number of rows but fewer columns. A common Primary Key links the tables.  
**Horizontal partitioning**: it divides a table into multiple smaller tables. Each table is a separate data store, and it contains the same number of columns, but fewer rows.

3. Create two tables - categorical and numeric.  
   categorical(Student_ID (PK), School, Sex, Age, Home_state, Home_town, Mother_tongue)  
   numeric(Student_ID (PK), Height_cm, Weight, BMI)  
   Declare data types and CHECKS as appropriate in both the schemas.  
   Partitioning the all_students data vertically, insert into the appropriate tables.  
   With a query, setting relationship on PK, display the first five rows of data gathered from the two tables.   

```python
sql_create_table = """
CREATE TABLE if NOT exists categorical(
    Student_ID TEXT PRIMARY KEY,
    School TEXT,
    Sex TEXT CHECK(Sex IN ('M', 'F')),
    Age INTEGER CHECK(Age < 100),
    Home_Town TEXT,
    Home_State TEXT,
    Mother_Tongue TEXT
);
"""
con.execute(sql_create_table)
```

```python
sql_insert_data = """
INSERT OR IGNORE INTO categorical (Student_ID, School, Sex, Age, Home_Town, Home_State, Mother_Tongue)
SELECT Student_ID, School, Sex, Age, Home_Town, Home_State, Mother_Tongue
FROM all_students;
"""
con.execute(sql_insert_data)
```

```python
con.commit()
```

```python
pd.read_sql("select * from categorical",con)
```

```python
con.execute("DROP TABLE IF EXISTS numerical")
sql = """
    CREATE TABLE numerical (
    Student_ID TEXT PRIMARY KEY,
    Height_cm REAL CHECK (Height_cm BETWEEN 100 AND 200),
    Weight_Kg REAL CHECK (Weight_Kg BETWEEN 10 AND 300),
    BMI REAL CHECK (BMI BETWEEN 5 AND 50)
);"""

con.execute(sql)
```

```python
sql_insert_data = """
INSERT INTO numerical (Student_ID, Height_cm,Weight_Kg,BMI )
SELECT Student_ID, Height_cm,Weight_Kg,BMI
FROM all_students;
"""

```

```python
con.execute(sql_insert_data)
```

```python
pd.read_sql("SELECT * FROM numerical",con)
```
